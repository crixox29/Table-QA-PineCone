```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os

def leer_y_procesar_archivo(file_path):
    with open(file_path, 'r', encoding='ISO-8859-1') as file:
        lines = file.readlines()

    start_index = None
    for i, line in enumerate(lines):
        if "T/sec" in line:
            start_index = i
            break

    if start_index is not None:
        data_lines = lines[start_index + 1:]
        data = [line.strip().split("\t") for line in data_lines if line.strip()]

        filtered_data = []
        for row in data:
            if len(row) >= 3:  # Asegurarse de que hay al menos 3 elementos
                # Eliminar los dos primeros caracteres de "Step" y asegurarse de que la fila tenga suficientes elementos
                filtered_data.append([row[0], row[1], row[-1][2:]])

        if filtered_data:
            df = pd.DataFrame(filtered_data, columns=["T/sec", "C1/°C", "Step"])
            df["T/sec"] = df["T/sec"].str.replace(",", ".").astype(float)
            return df
    return pd.DataFrame()



folder_path = '/content/IBL_2/'  # Ajusta este camino a tu carpeta específica
files = glob.glob(os.path.join(folder_path, "*.txt"))

# DataFrame para acumular las diferencias de todos los archivos
all_differences = pd.DataFrame()

for file in files:
    df = leer_y_procesar_archivo(file)
    if not df.empty:
        grouped = df.groupby('Step')['T/sec'].agg(['max', 'min'])
        grouped['difference'] = grouped['max'] - grouped['min']
        grouped['Step'] = grouped.index  # Convertir el índice 'Step' en una columna
        all_differences = pd.concat([all_differences, grouped[['Step', 'difference']].reset_index(drop=True)], ignore_index=True)

# Visualizar todas las diferencias acumuladas en un diagrama de caja
plt.figure(figsize=(12, 8))
sns.boxplot(x='Step', y='difference', data=all_differences)
plt.title('Distribución de Diferencias de Tiempo por Step - Todos los Archivos')
plt.xlabel('Step')
plt.ylabel('Diferencia de Tiempo (T/sec)')
plt.xticks(rotation=90)
plt.show()
```
